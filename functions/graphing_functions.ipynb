```python
# Importing libraries
import matplotlib.pyplot as plt
import numpy as np
import math as m
```

```python
def graph(formula, x_range):
    x = np.array(x_range)
    y = formula(x)
    # Handle values near x = 0 where tan is vanishingly small and cotan undefined or approaching infinity
    # Because arange with float generates a tiny near-zero value rather than actual 0, this generates a very large cotan value
    # The true graph would indeed have a vertical asymptote near x = 0, but showing even an approximation of it would make the graph unreadable due to scaling
    # Alternative would be limit y at e.g. 50 to illulstrate the asymptote.
    y[np.isclose(x, 0, atol=1e-10)] = np.nan
    plt.plot(x, y)
    plt.show()
    for i in zip(x,y):
        print(i)

graph(lambda x: 1/np.tan(x), np.arange(-5,5, 0.1))
```
